Usando el metodo del codo para encontrar el numero de cluster adecuado

```python
from sklearn.impute import SimpleImputer
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Assuming 'X' is your data with NaN values
imputer = SimpleImputer(strategy='mean')  # or strategy='median', 'most_frequent'
X_imputed = imputer.fit_transform(X)

inertia = []
k_values = range(2, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_imputed)  # Use the imputed data here
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(12, 5))
plt.plot(k_values, inertia, marker='o')
plt.show()
```

Entrenamos y usamos el Kmeans para separar en 3 grupos

```python
k_optimo = 3
kmeans = KMeans(n_clusters=k_optimo, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X)
```

usamos PCA para obtener dos componentes principales

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
plt.figure(figsize=(10, 6))

scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap="viridis", alpha=0.6)
plt.colorbar(scatter, label="Grupos")
plt.show()
```

```python
# prompt: genera la probabilidad que queder herido y probabilidad de muerte en un accidente

# Probabilidad de heridos y muertos por tipo de accidente y cluster

for cluster in range(3):
  cluster_df = df[df['Cluster'] == cluster]
  total_accidentes_cluster = len(cluster_df)

  for tipo_accidente in [2, 10, 11]: # Peatón, Vehículo, Ciclista
    accidentes_tipo = cluster_df[cluster_df['TIPACCID'] == tipo_accidente]
    num_accidentes_tipo = len(accidentes_tipo)

    if total_accidentes_cluster > 0:
      probabilidad_accidente = (num_accidentes_tipo / total_accidentes_cluster) * 100
      print(f"Cluster {cluster}, Tipo de Accidente {tipo_accidente}:")
      print(f"- Probabilidad del tipo de accidente: {probabilidad_accidente:.2f}%")

      # Probabilidad de heridos y muertos (asumiendo que TOTMUERTOS y TOTHERIDOS son válidas)
      if not accidentes_tipo.empty:
          total_muertos = accidentes_tipo['TOTMUERTOS'].sum()
          total_heridos = accidentes_tipo['TOTHERIDOS'].sum()

          if num_accidentes_tipo > 0:
              prob_muerte = (total_muertos / num_accidentes_tipo) * 100
              prob_heridos = (total_heridos / num_accidentes_tipo) * 100
              print(f"- Probabilidad de muerte: {prob_muerte:.2f}%")
              print(f"- Probabilidad de heridos: {prob_heridos:.2f}%")
          else:
              print("No hay accidentes de este tipo para calcular probabilidad de muerte o heridos")
    else:
      print(f"No hay accidentes en el cluster {cluster}")
    print("---")
```
