```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_nacional.csv", encoding='latin-1')
```

tipos de accidente son más comunes en cierto estado del pavimento.

```python
df['CAPAROD'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Frecuencia de tipos de pavimento (CAPAROD)")
plt.show()
```

Visualización de Accidentes Fatales por Tipo de Pavimento.

```python
# Ejemplo: TOTMUERTOS promedio según URBANA
print(df.groupby('URBANA')['TOTMUERTOS'].mean())

# O un crosstab con rangos de TOTMUERTOS
# Primero creamos una categoría binaria o por rangos de TOTMUERTOS
df['FATAL'] = (df['TOTMUERTOS'] > 0).astype(int)
pd.crosstab(df['URBANA'], df['FATAL'])
```

```python
print("\nDistribución de accidentes fatales por tipo de pavimento:")
fatal_by_pavement = pd.crosstab(df['CAPAROD'], df['FATAL'])
print(fatal_by_pavement)
```

```python
plt.figure(figsize=(8, 5))
fatal_by_pavement.plot(kind='bar', stacked=True, color=['lightcoral', 'lightgreen'])
plt.title("Accidentes Fatales por Tipo de Pavimento")
plt.xlabel("Tipo de Pavimento (CAPAROD)")
plt.ylabel("Número de Accidentes")
plt.show()
```

Visualización de Accidentes Fatales por Tipo de Vía Urbana/Suburbana

```python
# Accidentes fatales por vía urbana
plt.figure(figsize=(8, 5))
df.groupby('URBANA')['FATAL'].sum().plot(kind='bar', color='tomato')
plt.title("Accidentes Fatales según el Tipo de Vía Urbana")
plt.xlabel("Tipo de Vía Urbana")
plt.ylabel("Número de Accidentes Fatales")
plt.show()

# Accidentes fatales por vía suburbana
plt.figure(figsize=(8, 5))
df.groupby('SUBURBANA')['FATAL'].sum().plot(kind='bar', color='darkcyan')
plt.title("Accidentes Fatales según el Tipo de Vía Suburbana")
plt.xlabel("Tipo de Vía Suburbana")
plt.ylabel("Número de Accidentes Fatales")
plt.show()
```

```python
# Crear una variable de clasificación según la hora del día
def clasificar_hora(hora):
    if 0 <= hora < 6:
        return 'Madrugada'
    elif 6 <= hora < 12:
        return 'Mañana'
    elif 12 <= hora < 18:
        return 'Tarde'
    else:
        return 'Noche'

# Crear la columna 'HORA_DEL_DIA' a partir de la hora registrada
df['HORA_DEL_DIA'] = df['HORA'].apply(clasificar_hora)

# Crear la columna 'PERIODO_DIA' según el día de la semana
df['PERIODO_DIA'] = df['DIASEMANA'].apply(lambda x: 'Laboral' if x <= 5 else 'Fin de Semana')

# Crear la columna 'FECHA_COMPLETA' combinando año, mes y día
df['FECHA_COMPLETA'] = pd.to_datetime(df[['ANIO', 'MES', 'DIA']].rename(columns={
    'ANIO': 'year',
    'MES': 'month',
    'DIA': 'day'
}))

# Verificar el resultado
print(df[['HORA', 'HORA_DEL_DIA', 'DIASEMANA', 'PERIODO_DIA', 'FECHA_COMPLETA']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Frecuencia de accidentes según el período del día
periodo_counts = df['HORA_DEL_DIA'].value_counts()

# Gráfico de barras
plt.figure(figsize=(8, 5))
sns.barplot(x=periodo_counts.index, y=periodo_counts.values, palette='viridis')
plt.title("Frecuencia de Accidentes según el Período del Día")
plt.xlabel("Período del Día")
plt.ylabel("Número de Accidentes")
plt.show()

# Mostrar los resultados numéricos
print("Frecuencia de accidentes por período del día:")
print(periodo_counts)
```

Frecuencia de Accidentes en Días Laborales vs Fines de Semana

```python
# Frecuencia de accidentes según el período del día
dia_counts = df['PERIODO_DIA'].value_counts()

# Gráfico de barras
plt.figure(figsize=(8, 5))
sns.barplot(x=dia_counts.index, y=dia_counts.values, palette='coolwarm')
plt.title("Frecuencia de Accidentes en Días Laborales vs Fines de Semana")
plt.xlabel("Período del Día")
plt.ylabel("Número de Accidentes")
plt.show()

# Mostrar los resultados numéricos
print("Frecuencia de accidentes en días laborales vs fines de semana:")
print(dia_counts)
```

```python
# Convertir la columna de fecha en índice
df.set_index('FECHA_COMPLETA', inplace=True)

# Resumir el número de accidentes por día
accidentes_por_fecha = df.resample('D').size()

# Gráfico de líneas para analizar la tendencia temporal
plt.figure(figsize=(12, 6))
plt.plot(accidentes_por_fecha, color='blue')
plt.title("Evolución de Accidentes a lo Largo del Tiempo")
plt.xlabel("Fecha")
plt.ylabel("Número de Accidentes")
plt.show()

# Mostrar los primeros resultados
print("Frecuencia de accidentes en fechas específicas:")
print(accidentes_por_fecha.head(20))
```
