```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_nacional.csv", encoding='latin-1')
```

tipos de accidente son m√°s comunes en cierto estado del pavimento.

```python
df['CAPAROD'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Frecuencia de tipos de pavimento (CAPAROD)")
plt.show()
```

Visualizaci√≥n de Accidentes Fatales por Tipo de Pavimento.

```python
# Ejemplo: TOTMUERTOS promedio seg√∫n URBANA
print(df.groupby('URBANA')['TOTMUERTOS'].mean())

# O un crosstab con rangos de TOTMUERTOS
# Primero creamos una categor√≠a binaria o por rangos de TOTMUERTOS
df['FATAL'] = (df['TOTMUERTOS'] > 0).astype(int)
pd.crosstab(df['URBANA'], df['FATAL'])
```

```python
print("\nDistribuci√≥n de accidentes fatales por tipo de pavimento:")
fatal_by_pavement = pd.crosstab(df['CAPAROD'], df['FATAL'])
print(fatal_by_pavement)
```

```python
plt.figure(figsize=(8, 5))
fatal_by_pavement.plot(kind='bar', stacked=True, color=['lightcoral', 'lightgreen'])
plt.title("Accidentes Fatales por Tipo de Pavimento")
plt.xlabel("Tipo de Pavimento (CAPAROD)")
plt.ylabel("N√∫mero de Accidentes")
plt.show()
```

Visualizaci√≥n de Accidentes Fatales por Tipo de V√≠a Urbana/Suburbana

```python
# Accidentes fatales por v√≠a urbana
plt.figure(figsize=(8, 5))
df.groupby('URBANA')['FATAL'].sum().plot(kind='bar', color='tomato')
plt.title("Accidentes Fatales seg√∫n el Tipo de V√≠a Urbana")
plt.xlabel("Tipo de V√≠a Urbana")
plt.ylabel("N√∫mero de Accidentes Fatales")
plt.show()

# Accidentes fatales por v√≠a suburbana
plt.figure(figsize=(8, 5))
df.groupby('SUBURBANA')['FATAL'].sum().plot(kind='bar', color='darkcyan')
plt.title("Accidentes Fatales seg√∫n el Tipo de V√≠a Suburbana")
plt.xlabel("Tipo de V√≠a Suburbana")
plt.ylabel("N√∫mero de Accidentes Fatales")
plt.show()
```

```python
# Crear una variable de clasificaci√≥n seg√∫n la hora del d√≠a
def clasificar_hora(hora):
    if 0 <= hora < 6:
        return 'Madrugada'
    elif 6 <= hora < 12:
        return 'Ma√±ana'
    elif 12 <= hora < 18:
        return 'Tarde'
    else:
        return 'Noche'

# Crear la columna 'HORA_DEL_DIA' a partir de la hora registrada
df['HORA_DEL_DIA'] = df['HORA'].apply(clasificar_hora)

# Crear la columna 'PERIODO_DIA' seg√∫n el d√≠a de la semana
df['PERIODO_DIA'] = df['DIASEMANA'].apply(lambda x: 'Laboral' if x <= 5 else 'Fin de Semana')

# Crear la columna 'FECHA_COMPLETA' combinando a√±o, mes y d√≠a
df['FECHA_COMPLETA'] = pd.to_datetime(df[['ANIO', 'MES', 'DIA']].rename(columns={
    'ANIO': 'year',
    'MES': 'month',
    'DIA': 'day'
}))

# Verificar el resultado
print(df[['HORA', 'HORA_DEL_DIA', 'DIASEMANA', 'PERIODO_DIA', 'FECHA_COMPLETA']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Frecuencia de accidentes seg√∫n el per√≠odo del d√≠a
periodo_counts = df['HORA_DEL_DIA'].value_counts()

# Gr√°fico de barras
plt.figure(figsize=(8, 5))
sns.barplot(x=periodo_counts.index, y=periodo_counts.values, palette='viridis')
plt.title("Frecuencia de Accidentes seg√∫n el Per√≠odo del D√≠a")
plt.xlabel("Per√≠odo del D√≠a")
plt.ylabel("N√∫mero de Accidentes")
plt.show()

# Mostrar los resultados num√©ricos
print("Frecuencia de accidentes por per√≠odo del d√≠a:")
print(periodo_counts)
```

Frecuencia de Accidentes en D√≠as Laborales vs Fines de Semana

```python
# Frecuencia de accidentes seg√∫n el per√≠odo del d√≠a
dia_counts = df['PERIODO_DIA'].value_counts()

# Gr√°fico de barras
plt.figure(figsize=(8, 5))
sns.barplot(x=dia_counts.index, y=dia_counts.values, palette='coolwarm')
plt.title("Frecuencia de Accidentes en D√≠as Laborales vs Fines de Semana")
plt.xlabel("Per√≠odo del D√≠a")
plt.ylabel("N√∫mero de Accidentes")
plt.show()

# Mostrar los resultados num√©ricos
print("Frecuencia de accidentes en d√≠as laborales vs fines de semana:")
print(dia_counts)
```

```python
# Convertir la columna de fecha en √≠ndice
df.set_index('FECHA_COMPLETA', inplace=True)

# Resumir el n√∫mero de accidentes por d√≠a
accidentes_por_fecha = df.resample('D').size()

# Gr√°fico de l√≠neas para analizar la tendencia temporal
plt.figure(figsize=(12, 6))
plt.plot(accidentes_por_fecha, color='blue')
plt.title("Evoluci√≥n de Accidentes a lo Largo del Tiempo")
plt.xlabel("Fecha")
plt.ylabel("N√∫mero de Accidentes")
plt.show()

# Mostrar los primeros resultados
print("Frecuencia de accidentes en fechas espec√≠ficas:")
print(accidentes_por_fecha.head(20))
```
