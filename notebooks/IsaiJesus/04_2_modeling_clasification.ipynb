# Classification

Para el caso de la clasificación, la variable a predecir será `CAUSAACCI`, a partir de otras variables que se encuentran mayormente correlacionadas entre sí, otras que se eligieron a criterio.
La forma para la elección del modelo será con el mejor resultado dado por **Cross Validation** con diferentes hiperparámetros aplicados a cada modelo, tomando los modelos de clasificación como **Decision Tree, Random Forest, LightGBM y CatBoost**.

`CAUSAACCI` (Causa probable o presunta del accidente)
- 1: Conductor
- 2: Peatón o pasajero
- 3: Falla de vehículo
- 4: Mala condición del camino
- 5: Otra

Además, algo importante es que debido a que los datos del dataset están muy desbalanceados, se mostrará el caso con el dataset completo pero evaluado con **F1-score** y el caso de hacer un **submuestreo** para tener el dataset más balanceado.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, RandomizedSearchCV, StratifiedKFold, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv('../data/processed/processed_nacional.csv')
```

```python
df.shape
```

```python
df.columns
```

```python
y = df['CAUSAACCI']

X = df[[
    'EDO', 'MES', 'HORA', 'MINUTOS', 'DIA', 'DIASEMANA', 'ZONA',
    'AUTOMOVIL', 'CAMPASAJ', 'MOTOCICLET', 'BICICLETA', 'TRANPUBLICO',
    'VEHICARGA', 'OTROVEHIC', 'SEXO', 'EDAD', 'CAPAROD', 'ALIENTOCINT',
    'TOTMUERTOS', 'TOTHERIDOS', 'CLASE', 'TIPACCID'
]]

print(y.shape, X.shape)
```

```python
X_train, X_test, y_train, y_test, = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X_train)
```

```python
# Asignación de variables a modelos

tree = DecisionTreeClassifier()
r_forest = RandomForestClassifier()
lgbm = LGBMClassifier()
catboost = CatBoostClassifier(verbose=0)
```

## Dataset completo evaluado con F1-score

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

### 1. Decision Tree Classifier

```python
tree_params = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 7, 9, 10, 12, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 3, 5],
    'splitter': ['best', 'random'],
    'max_features': ['sqrt', 'log2', None],
    'class_weight': [None, 'balanced'],
    'min_impurity_decrease': [0.0, 0.05, 0.1],
    'random_state': [42]
}
```

```python
tree_cv = RandomizedSearchCV(
    estimator=tree,
    param_distributions=tree_params,
    cv=cv,
    scoring='f1_weighted',
    verbose=1
)
```

```python
tree_cv.fit(X_scaled, y_train)
```

```python
tree_cv.best_params_
```

```python
tree_cv.best_score_
```

## Submuestreo para balancear los datos

```python
cv = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
rus = RandomUnderSampler(random_state=42, sampling_strategy='auto')
```

```python
X_res, y_res = rus.fit_resample(X_scaled, y_train.values.ravel())
```

# Prueba de Test con LightGBM

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

## Dataset completo

```python
X_test_scaled = scaler.transform(X_test)
```

```python
best_model = LGBMClassifier(**lgbm_cv.best_params_, random_forest=42)
```

```python
best_model.fit(X_scaled, y_train)
```

```python
y_pred = best_model.predict(X_test_scaled)
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1-score: {f1:.4f}')
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
class_labels = {
    1: "Conductor",
    2: "Peatón o pasajero",
    3: "Falla de vehículo",
    4: "Mala condición del camino",
    5: "Otro"
}

labels = [class_labels[i] for i in np.unique(y_test)]

plt.figure(figsize=(8, 6))

sns.heatmap(conf_matrix, annot=True, fmt="d", cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.title("Martiz de confusión")
plt.xlabel("Predicciones")
plt.ylabel("Valores reales")

plt.show()
```
