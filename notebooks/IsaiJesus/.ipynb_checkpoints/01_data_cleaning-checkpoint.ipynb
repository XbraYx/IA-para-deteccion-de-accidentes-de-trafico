# Data Cleaning

## Dataset nacional

```python
df[['CALLE1', 'CALLE2', 'CARRETERA', 'LATITUD', 'LONGITUD']].head()
```

```python
df = df.drop('CARRETERA', axis=1)
```

```python
calle1_lost = df[df['CALLE1'].isnull()][['CALLE1', 'CALLE2', 'LONGITUD', 'LATITUD']]
idx_null = df[df['CALLE1'].isnull()].index

print("Registros sin valores en CALLE1")
calle1_lost
```

```python
from geopy.geocoders import Nominatim
from time import sleep
import re

calle1_lost[['LATITUD', 'LONGITUD']]

geolocator = Nominatim(user_agent="geoapi")

regex = r"([^,]+), [^,]+, Municipio de ([^,]+), ([^,]+),"
locations = []

for index, row in calle1_lost.iterrows():
    lat = row['LATITUD']
    lon = row['LONGITUD']
    
    try:
        location = geolocator.reverse((lat, lon), language="es")
        direccion = location.address if location else "No se encontró información"
        # print(direccion + '\n')
        match = re.search(regex, direccion)
        if match:
            street = match.group(1).strip()
            locations.append(street)
        else:
            locations.append("Calle sin nombre")
        
        sleep(1)
    except Exception as e:
        print(f"Error con las coordenadas ({lat}, {lon}): {e}")
        
# Asignar los valores de 'locations' a índices en la columna 'CALLE1' sin valores
df.loc[idx_null, 'CALLE1'] = locations
```

Para encontrar la edad de los lesionados/occisos se puede hacer a partir de los rangos de edad y del total de lesionados/occisos.
En el caso de los occisos, la mayoría de los datos faltantes deben ser 0, ya que en esos casos el total de occisos es de 0, por eso esos valores son nulos.
Para ambos casos, es objeto ya que en algunos casos hay más de una edad.
El dato "SD" en algunas columnas significa **Se desconoce**.

```python
df2.loc[df2["total_occisos"] == 0, "edad_occiso"] = 0
```
